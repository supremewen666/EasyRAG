# Build A Tiny Eval Set

This notebook is the initial scaffold for the evaluation stage. It is intentionally compact, but it already defines the teaching contract for the section and gives the reader a clear next step.

## What you will do

- build a tiny but inspectable set of queries, gold evidence, and expected answers
- inspect the artifacts that make the metric or judgment legible
- record the failure mode before proposing a fix

## Why this matters

A small eval set is easier to trust and debug than a large opaque benchmark when you are still learning the system.
